```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
X, Y = np.meshgrid(x, y)

# Simulate a Gaussian beam profile
w0 = 1  # Beam waist
I0 = 1  # Peak intensity
beam_intensity = I0 * np.exp(-(X**2 + Y**2) / w0**2)

# Plot the beam intensity
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, beam_intensity, cmap="inferno")
plt.colorbar(label="Intensity")
plt.xlabel("X Position (mm)")
plt.ylabel("Y Position (mm)")
plt.title("SLM Laser Beam Intensity Profile")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
w0 = 0.05  # Beam waist (radius)
I0 = 1.0   # Peak intensity

# Create a grid
x = np.linspace(-0.1, 0.1, 400)
y = np.linspace(-0.1, 0.1, 400)
X, Y = np.meshgrid(x, y)

# Gaussian intensity profile
I = I0 * np.exp(-2 * (X**2 + Y**2) / w0**2)

# Plot the intensity profile
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, I, cmap="inferno")
plt.colorbar(label="Intensity (W/mÂ²)")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("Gaussian Laser Beam Intensity Profile")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define laser path
x = np.linspace(0, 5, 50)  # Laser moves from x=0 to x=5
y = np.zeros_like(x)       # No movement in Y-direction

# Plot the laser path
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'r-', linewidth=2, label="Laser Path")
plt.scatter(x[0], y[0], color="blue", label="Start")
plt.scatter(x[-1], y[-1], color="green", label="End")

# Labels and legend
plt.xlabel("X Position (mm)")
plt.ylabel("Y Position (mm)")
plt.title("Laser Scanning Path")
plt.legend()
plt.grid()
plt.show()
```
